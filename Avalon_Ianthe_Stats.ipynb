# Avalon Data Analytics
This iPython notebook analyses the data in the Avalon games played by my group of friends and me.

Rules of the game can be read [here](http://upload.snakesandlattes.com/rules/r/ResistanceAvalon.pdf).

## Importing and cleaning data
Data is stored game by game. Each Avalon game consists of different players, each taking on different characters. A player's character is hidden from the rest of the players. For the good side, a game can be won by winning three quests **and** avoiding the Merlin guess at the end of the game. For the bad side, a game can be won by winning three quests **or** getting the guessing Merlin at the end of the game.

Good side:  
* Helmet (actual term is Loyal Servant of Arthur)
* Percival
* Merlin

Bad side:  
* Minion (actual term is Minion of Modred)
* Morgana
* Modred
* Oberon


There are a maximum of five quests, with "1" representing a successful quest and "0" representing a quest that has failed.

## Sorting the Data
Sort data by players, then game type, then characters.

```python
def is_good(role):
    if (role == 'Helmet' or role == 'Merlin' or role == 'Percival'):
        return True
    else:
        return False

def is_win(role, victory):
    if is_good(role) and victory == 1:
        return True
    elif not is_good(role) and victory == 0:
        return True
    else:
        return False
```

```python
players = {}

dict_5 = five_people.to_dict('records')
dict_6 = six_people.to_dict('records')
dict_7 = seven_people.to_dict('records')
dict_8 = eight_people.to_dict('records')
dict_9 = nine_people.to_dict('records')
dict_10 = ten_people.to_dict('records')
all_dicts = [dict_5, dict_6, dict_7, dict_8, dict_9, dict_10]

five_characters = five_people.columns.tolist()[1:5]
six_characters = six_people.columns.tolist()[1:5]
seven_characters = seven_people.columns.tolist()[1:7]
eight_characters = eight_people.columns.tolist()[1:7]
nine_characters = nine_people.columns.tolist()[1:7]
ten_characters = ten_people.columns.tolist()[1:8]
characters = [five_characters, six_characters, seven_characters, eight_characters,
             nine_characters, ten_characters]

num_of_players = ["5_stats", "6_stats", "7_stats", "8_stats", "9_stats", "10_stats"]

# Generate dictionary with this structure: player -> game_type -> character -> play_count/wins/win_rate
for idx in range(0, 6):
    current_dict = all_dicts[idx]
    current_chars = characters[idx]
    current_stats = {}
    for game in current_dict:
        for character in current_chars:
            win = is_win(character, game['Victory'])
            count = 0
            if win:
                count = 1
            if type(game[character]) == str:
                for player in game[character].split(", "):
                    if player not in current_stats:
                        current_stats[player] = { 'wins': count, 'game_count': 1, character: 
                                                 {"win_rate": count, "wins": count, "play_count": 1} }
                    else:
                        current_stats[player]['game_count'] += 1
                        current_stats[player]['wins'] += count
                        if character not in current_stats[player]:
                            current_stats[player][character] = {"play_count": 1, "wins": count, "win_rate": count}
                        else:
                            current_stats[player][character]["play_count"] += 1
                            current_stats[player][character]["wins"] += count
                            current_stats[player][character]["win_rate"] = current_stats[player][character]["wins"] / current_stats[player][character]["play_count"] 

    for player, records in current_stats.items():
        records['win_rate'] = records['wins'] / records['game_count'] * 100
        if player not in players:
            players[player] = {}
        players[player][num_of_players[idx]] = records
```

```python
# Calculate overall stats for each game type and player.
for player, stats in players.items():
    total_game_count = 0
    weighted_sum = 0
    char_stats = {}
    stats["Overall"] = {}
    for stat_key, stat_value in stats.items():
        if stat_key == "Overall":
            break
        total_game_count += stat_value["game_count"]
        weighted_sum += stat_value["game_count"] * stat_value["win_rate"]
        for char_name, char_stats in stat_value.items():
            if char_name == "game_count" or char_name == "wins" or char_name == "win_rate":
                continue
            if char_name not in stats["Overall"]:
                stats["Overall"][char_name] = {"play_count": 0, "wins": 0}
            stats["Overall"][char_name]["play_count"] += char_stats["play_count"]
            stats["Overall"][char_name]["wins"] += char_stats["wins"]
    stats["Overall"]["win_rate"] = weighted_sum / total_game_count
    stats["Overall"]["wins"] = weighted_sum / 100
    stats["Overall"]["game_count"] = total_game_count
    for char_name, char_stats in stats["Overall"].items():
        if char_name == "game_count" or char_name == "wins" or char_name == "win_rate":
                continue
        char_stats["win_rate"] = char_stats["wins"] / char_stats["play_count"]

# Organises players alphabetically
sorted_players = {}
for player in sorted(players):
    sorted_players[player] = players[player]

# Prints giant dictionary
# for player, stat in players.items():
#     print(player)
#     for game_type, game_stats in stat.items():
#         print(game_type, game_stats)
#         print()
#     print("\n")
```

```python
# Generate nice tables with pandas for viewing of stats
for player, stat in sorted_players.items():
    print(player + "\n")
    for game_type, game_stats in stat.items():
        print(game_type)
        print(f"Win rate: {game_stats['win_rate']} ({game_stats['wins']}/{game_stats['game_count']})")
        game_stats_table = pd.DataFrame(game_stats).drop(columns=["game_count", "wins", "win_rate"]).T
        game_stats_table = game_stats_table[["win_rate", "wins", "play_count"]].reset_index()
        game_stats_table['index'] = pd.Categorical(game_stats_table['index'], ["Helmet", "Percival", "Merlin", "Minion", "Morgana", "Modred", "Oberon"])
        game_stats_table.sort_values("index", inplace=True)
        game_stats_table.rename(columns={"index": "Characters"}, inplace=True)
# WIP
#         game_stats_table.rename(columns=lambda x: x[1:])
        
        game_stats_table.rename(columns={"index": "Characters"}, inplace=True)
        game_stats_table.set_index("Characters", inplace=True)
        display(game_stats_table)
        print("\n")
    print()
```
